```python
import cv2
import numpy as np
```

```python
from keras.models import load_model
model = load_model('my model')
```

```python
face_cc = cv2.CascadeClassifier('C:/Users/Sumit/AppData/Local/Programs/Python/Python38/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
source = cv2.VideoCapture(0)             #start video captureq 
label_dict = {0:'NO MASK', 1:'MASK'}     #create label for image
color_dict = {0:(0,255,0), 1:(0,0,255)}  #create color dict for mask & no mask title
while(True):
  ret,video=source.read()                #start reading videos in frames (image)
  gray = cv2.cvtColor(video,cv2.COLOR_BGRA2GRAY)  #covert frame into black and white
  faces = face_cc.detectMultiScale(video,1.3,5)   #detect face and return parameters of facee

  for x,y,w,h in faces:
    face_img=gray[y:y+w,x:x+w]
    resized=cv2.resize(face_img,(100,100))        #resize frame
    normalized=resized/255.0
    reshaped=np.reshape(normalized,(1,100,100,1)) #convert 4d numpy array
    result=model.predict(reshaped)                #use model to predict mask
    
    label=np.argmax(result,axis=1)[0]             #get label of mask

    cv2.rectangle(video,(x,y),(x+w,y+h),color_dict[label],2)
    cv2.rectangle(video,(x,y-40),(x+w,y),color_dict[label],-1)
    cv2.putText(video,label_dict[label], (x,y-10), cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
  cv2.imshow('LIVE',video) #show video
  key=cv2.waitKey(1)       #get keypress from keyboard
  if(key==27):
    break                  #stop capturing video on ESC button press
cv2.destroyAllWindows()
source.release()
```
